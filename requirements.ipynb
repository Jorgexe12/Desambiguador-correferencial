# Dependencias
En este notebook solo existen dos comandos de pip:
- install: instala en el equipo todas las librerías almacenadas en el archivo requirements.txt.
- freeze: guarda todas las librerías del proyecto en el archivo requirements.txt.

Cabe destacar que hemos usado un entorno virtual, por lo que no sabemos con exactitud que pasa si se hace freeze fuera del entorno, ni si install las instala correctamente, en cualquier caso en el notebook al ejecutar el install te dice que elijas el kernel, por lo que debería funcionar sin entorno virtual.  
Por otro lado recomendamos no ejecutar al tuntún ya que podría haber esperas prolongadas debido al volumen de datos que manejamos, o podría fallar por alguna dependencia que falta al instalarse todo y perderse la salida. Nosotros vamos a ejecutar todo sin fallos, si se quiere ejecutar las celdas recomendamos tener un backup para poder volver fácilmente.
